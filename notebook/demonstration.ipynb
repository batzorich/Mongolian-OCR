```python
import os, sys, cv2

os.chdir("..")
sys.path.insert(0, os.getcwd())

from src.service import extract_text
```

```python
IMG_PATH = "test_imgs/img_001.jpg"
img = cv2.imread(IMG_PATH)
result = extract_text(img)
```

```python
print(result)
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import cv2

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig = plt.figure(figsize=(18, 8))
gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1.2])

ax0 = plt.subplot(gs[0])
ax0.imshow(img)
ax0.axis("on")
ax0.set_title("Input Image", fontsize=16, loc="left")

ax1 = plt.subplot(gs[1])
ax1.axis("off")
ax1.set_title("Recognized Text", fontsize=16, loc="left")
ax1.text(0, 1, result, fontsize=10, wrap=True, va='top', ha='left', transform=ax1.transAxes)

plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05, wspace=0.2)
plt.show()
```
